# How does the season impact posted wait times at Disney World?
*Posted* wait time: The amount of time Disney advertises you will wait on line


This analysis visualizes the average posted wait time for 4 Disney rides (one for each park).

```
# importing packages
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as st
import matplotlib.pyplot as plt
from google.colab import files
import warnings
warnings.filterwarnings("ignore")
```

```
# find datasets located on google drive
path = '/content/drive/My Drive/Data Science/Disney Data 2018/'
```

```
# import data
data = pd.read_csv(open(path + 'metadata_day_info.csv'))

# trim data columns
data = data[['DATE', 'DAYOFWEEK', 'MONTHOFYEAR', 'YEAR', 'SEASON', 'WDWSEASON']]
data.head()
```

```
# rename values
data['WDWSEASON'] = data['WDWSEASON'].str.replace('MARTIN LUTHER KING JUNIOR DAY','M.L.K. DAY')
data['SEASON'] = data['SEASON'].str.replace('MARTIN LUTHER KING JUNIOR DAY','M.L.K. DAY')

# rename columns
data.columns = ['Date', 'Day of Week', 'Month', 'Year', 'Season', 'Disney Season']

# set index
data = data.set_index('Date')
data.head()
```

# Import Ride Datasets
*Group posted wait times by date and compute average minutes waited*

```
# Magic Kingdom 
toyStory = pd.read_csv(open(path + 'ride_toystory.csv'))
toyStory = toyStory[toyStory['SPOSTMIN'] >= 0]
toyStory = toyStory.drop(columns=['datetime', 'SACTMIN'])
toyStory = toyStory.groupby(['date']).mean()
toyStory.columns = ['Toy Story Posted Wait']

# Epcot
soarin = pd.read_csv(open(path + 'ride_soarin.csv'))
soarin = soarin[soarin['SPOSTMIN'] >= 0]
soarin = soarin.drop(columns=['datetime', 'SACTMIN'])
soarin = soarin.groupby(['date']).mean()
soarin.columns = ['Soarin Posted Wait']

# Hollywood Studios
rockin = pd.read_csv(open(path + 'ride_rockincoaster.csv'))
rockin = rockin[rockin['SPOSTMIN'] >= 0]
rockin = rockin.drop(columns=['datetime', 'SACTMIN'])
rockin = rockin.groupby(['date']).mean()
rockin.columns = ['Rockin Posted Wait']

# Animal Kingdom
everest = pd.read_csv(open(path + 'ride_expeditioneverest.csv'))
everest = everest[everest['SPOSTMIN'] >= 0]
everest = everest.drop(columns=['datetime', 'SACTMIN'])
everest = everest.groupby(['date']).mean()
everest.columns = ['Everest Posted Wait']
```

```
# concatenate dataset and 4 ride dataframes into one working dataframe
combined = pd.concat([data, toyStory, soarin, rockin, everest], axis = 1)
combined = combined.drop(columns=['Day of Week', 'Month', 'Year', 'Season'])

# groupby "Disney Season" and compute average posted wait time in minutes
grouped = combined.groupby(["Disney Season"], as_index=False).mean()
grouped
```

# Visualizations

```
# Plot Toy Story posted wait times 
fig, ax = plt.subplots()
bar_width = 0.6
opacity = 0.9
error_config = {'ecolor': '0.3'}
rects1 = ax.bar(grouped['Disney Season'], grouped['Toy Story Posted Wait'], bar_width,
                alpha=opacity, color='gold')
plt.xticks(rotation=75)

ax.set_xlabel('Disney Season')
ax.set_ylabel('Average Posted Wait in Minutes')
ax.set_title('Toy Story Average Posted Wait Times by Disney Season')

fig.tight_layout()
plt.show()
plt.savefig('Toy Story Wait Times')
```

```
# Plot Soarin posted wait times
fig, ax = plt.subplots()
bar_width = 0.6
opacity = 0.6
error_config = {'ecolor': '0.3'}
rects1 = ax.bar(grouped['Disney Season'], grouped['Soarin Posted Wait'], bar_width,
                alpha=opacity, color='b')
plt.xticks(rotation=75)

ax.set_xlabel('Disney Season')
ax.set_ylabel('Average Posted Wait in Minutes')
ax.set_title('Soarin Average Posted Wait Times by Disney Season')

fig.tight_layout()
plt.show()
plt.savefig('Soarin Wait Times.png')
```
